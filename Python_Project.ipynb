```python
X = data.iloc[:,:].values
X
```

```python
#Taking care of missing data

#from sklearn.preprocessing import Imputer is outdated so used Simple Imputer
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = 'nan',strategy='constant',fill_value='No')
imputer = imputer.fit(X[:,21:22])
X[:,21:22] = imputer.transform(X[:,21:22])
X
```

```python
# replacing na values in college with No college 
data["Returned"].fillna("No", inplace = True) 
  
data
```

```python
data=data.drop(['Row ID','Order Date','Ship Date','Ship Mode','Customer Name','Segment','Country','City','State','Region','Product Name'],axis=1)
data
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_data = LabelEncoder()
data['Order ID']= labelencoder_data.fit_transform(data['Order ID'])
data['Customer ID']= labelencoder_data.fit_transform(data['Customer ID'])
data['Postal Code']= labelencoder_data.fit_transform(data['Postal Code'])
data['Product ID']= labelencoder_data.fit_transform(data['Product ID'])
data['Category']= labelencoder_data.fit_transform(data['Category'])
data['Sub-Category']= labelencoder_data.fit_transform(data['Sub-Category'])
data['Returned']= labelencoder_data.fit_transform(data['Returned'])
data
```

```python
#InDpendant Variable
X = data.iloc[:,:-1].values
X
```

```python
#Dpendant Variable
y = data.iloc[:,10:11].values #to confirm which column has NaN values
y
```

```python
#splitting the dataset into the training set and test data
#from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
# 60-80 training 20-30 testing
#random state 0 - same date will be taken as input to diff models
X_test
```

```python
#Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
```

```python
y_Pred = classifier.predict(X_test)
y_Pred
```

```python
#making the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_Pred)
cm
```

```python
from sklearn.metrics import precision_recall_fscore_support
precision_recall_fscore_support(y_test,y_Pred)
#Precision
#Recall
#fscore
#Support howmany positive & negative predicted count
#1st column for positive 2nd column for negative
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_Pred)
#this accuracy is same as we calculated by above formula
```
